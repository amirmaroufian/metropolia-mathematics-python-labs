### My solution

```python
x = np.linspace(0, 10, 500)

f = np.sqrt(2.7 * x)
g = (x - 5.24)**2

plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.title("Overview: Both functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')

plt.xlim(3.0, 4.0)
plt.ylim(2, 5)

plt.title("Root 1 zoom")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')

plt.xlim(7.0, 8.5)
plt.ylim(3, 6)

plt.title("Root 2 zoom")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
